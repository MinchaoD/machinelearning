```python
import pandas as pd
import numpy as np
```

```python

cleandf=pd.read_csv('cleandf.csv')

cleandf=cleandf.drop('Unnamed: 0', axis=1)
cleandf
```

```python
cleandf['Absenteeism Time in Hours'].median()
```

```python
cleandf['Absenteeism Time in Hours'].mean()
```

```python
#cleandf['Absenteeism Time in Hours']=cleandf['Absenteeism Time in Hours'].apply(lambda x: 0 if x<=cleandf['Absenteeism Time in Hours'].median() else 1)
cleandf['Absenteeism Time in Hours']=np.where(cleandf['Absenteeism Time in Hours']>cleandf['Absenteeism Time in Hours'].median(),1,0)
cleandf
```

```python
cleandf.columns.values
```

```python
cleandf=cleandf[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education', 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets', 'Absenteeism Time in Hours']]
cleandf
```

```python
y=cleandf['Absenteeism Time in Hours']
x=cleandf.drop('Absenteeism Time in Hours', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=101)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
dummies=x_train[['Reason_1','Reason_2','Reason_3','Reason_4','Education']]
to_scale=x_train[[ 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets',]]
```

```python
scaled_inputs=scaler.fit_transform(to_scale)
```

```python
scaled_inputs
```

```python
dummies=dummies.to_numpy()
```

```python
dummies.shape
```

```python
x_train=[]
x_train
```

```python
for i in range(dummies.shape[0]):
    x_train.append(np.concatenate((dummies[i],scaled_inputs[i])))
                   
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.coef_
```

```python
test_dummies=x_test[['Reason_1','Reason_2','Reason_3','Reason_4','Education']]
test_to_scale=x_test[[ 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets',]]
```

```python
test_scaled_inputs=scaler.transform(test_to_scale)
```

```python
test_dummies=test_dummies.to_numpy()
```

```python
x_test=[]
test_dummies.shape
```

```python
for i in range (test_dummies.shape[0]):
    x_test.append(np.concatenate((test_dummies[i],test_scaled_inputs[i])))
```

```python
pred=model.predict(x_test)
```
