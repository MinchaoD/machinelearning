```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('clean_no_dummy.csv')
df=df.drop('Unnamed: 0',axis=1)
```

```python
df['Absenteeism Time in Hours'].median()
```

```python
df['Absenteeism Time in Hours'].mean()
```

```python
#cleandf['Absenteeism Time in Hours']=cleandf['Absenteeism Time in Hours'].apply(lambda x: 0 if x<=cleandf['Absenteeism Time in Hours'].median() else 1)
df['Absenteeism Time in Hours']=np.where(df['Absenteeism Time in Hours']>df['Absenteeism Time in Hours'].median(),1,0)
df
```

```python
df.columns.values
```

```python
y=df['Absenteeism Time in Hours']
x=df.drop('Absenteeism Time in Hours', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=101)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_x_train=scaler.fit_transform(x_train)
scaled_x_train
```

```python
scaled_df=pd.DataFrame(columns=x.columns.values, data=scaled_x_train,index=x_train.index)
scaled_df
```

```python
def reason(x):
    if x==0:
        return 0
    elif 1<=x<=14:
        return 1
    elif 15<=x<=17:
        return 2
    elif 18<=x<=21:
        return 3
    else:
        return 4
```

```python
scaled_df['Reason for Absence']=x_train['Reason for Absence'].apply(reason)
scaled_df['Education']=x_train['Education']
scaled_df
```

```python
dummies=pd.get_dummies(scaled_df['Reason for Absence'],drop_first=True)
```

```python
dummies=dummies.add_prefix('Reason_')
```

```python
dummies
```

```python
dummies['Education']=scaled_df['Education'].apply(lambda x:  0 if x==1 else 1)
dummies
```

```python
scaled_df=scaled_df.drop('Education', axis=1)
```

```python
final_scaled=pd.concat([dummies,scaled_df],axis=1)
```

```python
final_scaled
```

```python
final_scaled=final_scaled.drop('Reason for Absence', axis=1)
```

```python
final_scaled
```

```python
x_train=np.array(final_scaled)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
summary_table=pd.DataFrame(columns=['Feature name'],data=final_scaled.columns.values)

summary_table
```

```python
summary_table['Coefficient'] = np.transpose(model.coef_)
summary_table=summary_table.sort_values('Coefficient')
summary_table
```

```python
summary_table['odds']=np.exp(summary_table.Coefficient)
```
